# Analisis exploratorio para variables categoricas y binarias 

Ahora vamos a ver la relacion de las variables categoricas y binarias con la variable 'Price_euros' que es la que nos interesa predecir a futuro y el analisis se va a ir orientado a preparar el terreno para poder predecir la misma

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df_laptops = pd.read_csv("dataset/p_laptops.csv", encoding='latin-1')
```

Usemos graficos de caja y bigotes para que mediante ellos ver la distribucion de la variable numerica 'Price_euros' a traves de las diferentes categorias que presentan las distintas variables categoricas que contiene el set de datos 

```python
# Definimos las columnas categricas para el anlisis (las que no son numricas continuas o binarias 0/1)
# Basndonos en el df.info() que tenamos y las transformaciones:
categorical_features = [
    'Company',
    'TypeName',
    'OS',
    'PrimaryStorageType',
    'SecondaryStorageType',
    'CPU_company',
    'GPU_company'
]

# Configuramos el estilo de los grficos
sns.set_style("whitegrid")
plt.figure(figsize=(18, 25)) # Aumentamos el tamao general de la figura

for i, col in enumerate(categorical_features):
    plt.subplot(len(categorical_features), 1, i + 1) # Creamos un subplot para cada caracterstica
    sns.boxplot(x=col, y='Price_euros', data=df_laptops, palette='viridis')
    plt.title(f'Distribucin de Precios por {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Precio (Euros)', fontsize=12)
    plt.xticks(rotation=45, ha='right') # Rotamos etiquetas para mejor legibilidad

plt.tight_layout() # Ajustamos el diseo para evitar superposiciones
plt.show()
```

Podemos usar tambien los graficos de caja y bigote para analizar tambien la relacion de 'Price_euros' con las columnas binarias (recordemos que estas anteriormente contenian valores Yes o No por lo cual tambien fuero variables categoricas en su momento)

```python
# Columnas binarias que tambin queremos analizar
binary_features = [
    'Touchscreen',
    'IPSpanel',
    'RetinaDisplay'
]

# Configuramos el estilo de los grficos
sns.set_style("whitegrid")
plt.figure(figsize=(12, 10)) # Ajustamos el tamao para 3 grficos

for i, col in enumerate(binary_features):
    plt.subplot(len(binary_features), 1, i + 1) # Creamos un subplot para cada caracterstica
    sns.boxplot(x=col, y='Price_euros', data=df_laptops, palette='Set2')
    plt.title(f'Distribucin de Precios por {col} (0=No, 1=Yes)', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Precio (Euros)', fontsize=12)
    plt.xticks(ticks=[0, 1], labels=['No', 'Yes']) # Etiquetas ms claras para 0 y 1

plt.tight_layout() # Ajustamos el diseo para evitar superposiciones
plt.show()
```

<center>Resumen de los graficos</center>

| Variable        | Observaciones del Grfico                                                                                                                                                                                                                                                                                                                                                                                                                          |
| :-------------- | :------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| `Touchscreen`   | - Las laptops con `Touchscreen=Yes` tienen una mediana de precio notablemente ms alta que aquellas con `Touchscreen=No`. <br> - La distribucin de precios para las pantallas tctiles (`Yes`) es ms compacta y se encuentra en un rango de precios superior, lo que sugiere que esta caracterstica se encuentra en modelos de gama media-alta o alta. <br> - Ambas categoras (`Yes` y `No`) presentan outliers en el extremo superior de precios, indicando que hay laptops muy caras con y sin pantalla tctil, pero la mayora de las laptops caras tienden a tenerla. |
| `IPSpanel`      | - Las laptops con `IPSpanel=Yes` tienen una mediana de precio considerablemente ms alta que aquellas con `IPSpanel=No`. <br> - La dispersin de precios es mayor en las laptops sin panel IPS (`No`), que cubren un rango de precios ms amplio desde los ms econmicos hasta los ms caros (con outliers). <br> - Las laptops con panel IPS (`Yes`) tienen una distribucin ms elevada y concentrada, sugiriendo que es una caracterstica premium.                  |
| `RetinaDisplay` | - Las laptops con `RetinaDisplay=Yes` (muy pocas, indicadas por la caja ms estrecha y los pocos puntos) muestran una mediana de precio significativamente ms alta en comparacin con las que no tienen esta caracterstica (`No`). <br> - La gran mayora de las laptops no tienen `RetinaDisplay=No`, y esta categora abarca un rango de precios muy amplio, desde los ms bajos hasta los ms altos. <br> - La presencia de `RetinaDisplay` se asocia fuertemente con un segmento de mercado de muy alto precio.                                                               |
