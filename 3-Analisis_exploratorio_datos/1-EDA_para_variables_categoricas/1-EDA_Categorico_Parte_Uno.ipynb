# Analisis exploratorio para variables categoricas y binarias 

Ahora vamos a ver la relacion de las variables categoricas y binarias con la variable 'Price_euros' que es la que nos interesa predecir a futuro y el analisis se va a ir orientado a preparar el terreno para poder predecir la misma

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df_laptops = pd.read_csv("dataset/p_laptops.csv", encoding='latin-1')
```

Usemos graficos de caja y bigotes para que mediante ellos ver la distribucion de la variable numerica 'Price_euros' a traves de las diferentes categorias que presentan las distintas variables categoricas que contiene el set de datos 

```python
# Definimos las columnas categóricas para el análisis (las que no son numéricas continuas o binarias 0/1)
# Basándonos en el df.info() que teníamos y las transformaciones:
categorical_features = [
    'Company',
    'TypeName',
    'OS',
    'PrimaryStorageType',
    'SecondaryStorageType',
    'CPU_company',
    'GPU_company'
]

# Configuramos el estilo de los gráficos
sns.set_style("whitegrid")
plt.figure(figsize=(18, 25)) # Aumentamos el tamaño general de la figura

for i, col in enumerate(categorical_features):
    plt.subplot(len(categorical_features), 1, i + 1) # Creamos un subplot para cada característica
    sns.boxplot(x=col, y='Price_euros', data=df_laptops, palette='viridis')
    plt.title(f'Distribución de Precios por {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Precio (Euros)', fontsize=12)
    plt.xticks(rotation=45, ha='right') # Rotamos etiquetas para mejor legibilidad

plt.tight_layout() # Ajustamos el diseño para evitar superposiciones
plt.show()
```

| `SecondaryStorageType` | - La mayoría de las laptops no tienen almacenamiento secundario (`No`), y este grupo muestra una mediana de precio en el rango medio-alto. <br> - Las laptops con **HDD** como almacenamiento secundario tienen precios medianos más bajos en comparación con las que no tienen (lo que podría indicar una configuración más básica o económica en el disco principal). <br> - El **SSD** como almacenamiento secundario se asocia con precios medianos más altos y un rango más amplio, indicando configuraciones duales de gama alta. <br> - **Hybrid** secundario tiene muy pocos datos y precios variados.                                         |
| `CPU_company`          | - **Intel** domina la distribución de precios, con una mediana en el rango medio y un rango muy amplio que incluye los precios más altos del dataset, debido a su mayor presencia. <br> - **AMD** tiene una mediana de precios más baja y un rango más estrecho, lo que sugiere que las laptops con CPU AMD tienden a ser más económicas en este dataset. <br> - **Samsung** (con solo 1 dato) es un outlier de precio extremadamente alto, lo cual no es representativo.                                                                                             |
| `GPU_company`          | - **Nvidia** se asocia con precios medianos más altos y un rango amplio, lo que es coherente con su presencia en laptops gaming y de alto rendimiento. <br> - **AMD** se sitúa en un rango de precios medio-bajo, con una mediana inferior a la de Nvidia. <br> - **Intel** (gráficos integrados) es la categoría más numerosa y con la mediana de precios más baja, lo cual es esperable ya que suelen estar en laptops de menor gama. <br> - **ARM** (con solo 1 dato) es un outlier de precio muy bajo, no representativo.                                                                          |

Podemos usar tambien los graficos de caja y bigote para analizar tambien la relacion de 'Price_euros' con las columnas binarias (recordemos que estas anteriormente contenian valores Yes o No por lo cual tambien fuero variables categoricas en su momento)

```python
# Columnas binarias que también queremos analizar
binary_features = [
    'Touchscreen',
    'IPSpanel',
    'RetinaDisplay'
]

# Configuramos el estilo de los gráficos
sns.set_style("whitegrid")
plt.figure(figsize=(12, 10)) # Ajustamos el tamaño para 3 gráficos

for i, col in enumerate(binary_features):
    plt.subplot(len(binary_features), 1, i + 1) # Creamos un subplot para cada característica
    sns.boxplot(x=col, y='Price_euros', data=df_laptops, palette='Set2')
    plt.title(f'Distribución de Precios por {col} (0=No, 1=Yes)', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Precio (Euros)', fontsize=12)
    plt.xticks(ticks=[0, 1], labels=['No', 'Yes']) # Etiquetas más claras para 0 y 1

plt.tight_layout() # Ajustamos el diseño para evitar superposiciones
plt.show()
```

<center>Resumen de los graficos</center>

| Variable        | Observaciones del Gráfico                                                                                                                                                                                                                                                                                                                                                                                                                          |
| :-------------- | :------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| `Touchscreen`   | - Las laptops con `Touchscreen=Yes` tienen una mediana de precio notablemente más alta que aquellas con `Touchscreen=No`. <br> - La distribución de precios para las pantallas táctiles (`Yes`) es más compacta y se encuentra en un rango de precios superior, lo que sugiere que esta característica se encuentra en modelos de gama media-alta o alta. <br> - Ambas categorías (`Yes` y `No`) presentan outliers en el extremo superior de precios, indicando que hay laptops muy caras con y sin pantalla táctil, pero la mayoría de las laptops caras tienden a tenerla. |
| `IPSpanel`      | - Las laptops con `IPSpanel=Yes` tienen una mediana de precio considerablemente más alta que aquellas con `IPSpanel=No`. <br> - La dispersión de precios es mayor en las laptops sin panel IPS (`No`), que cubren un rango de precios más amplio desde los más económicos hasta los más caros (con outliers). <br> - Las laptops con panel IPS (`Yes`) tienen una distribución más elevada y concentrada, sugiriendo que es una característica premium.                  |
| `RetinaDisplay` | - Las laptops con `RetinaDisplay=Yes` (muy pocas, indicadas por la caja más estrecha y los pocos puntos) muestran una mediana de precio significativamente más alta en comparación con las que no tienen esta característica (`No`). <br> - La gran mayoría de las laptops no tienen `RetinaDisplay=No`, y esta categoría abarca un rango de precios muy amplio, desde los más bajos hasta los más altos. <br> - La presencia de `RetinaDisplay` se asocia fuertemente con un segmento de mercado de muy alto precio.                                                               |
