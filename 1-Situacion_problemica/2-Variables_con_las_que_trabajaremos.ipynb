# Explicación de Variables del Dataset de Laptops

Este documento explica cada una de las columnas (variables) presentes en el dataset de precios de laptops, proporcionando una descripción clara en español para facilitar su comprensión y uso en análisis o modelos.

---

### **`Company`**
* **Descripción:** Representa la marca o fabricante de la laptop.
* **Ejemplo:** Apple, HP, Dell, Lenovo.
* **Tipo de dato:** Categórico (texto).

### **`TypeName`**
* **Descripción:** Describe el tipo o categoría general de la laptop, indicando su diseño y propósito principal.
* **Ejemplo:** Notebook, Gaming, Ultrabook, 2 in 1.
* **Tipo de dato:** Categórico (texto).

### **`Inches`**
* **Descripción:** Indica el tamaño diagonal de la pantalla de la laptop, medido en pulgadas.
* **Ejemplo:** 13.3, 15.6, 17.3.
* **Tipo de dato:** Numérico (flotante).

### **`Ram`**
* **Descripción:** Especifica la cantidad de memoria de acceso aleatorio (RAM) instalada en la laptop, medida en Gigabytes (GB). Es crucial para la multitarea y el rendimiento general.
* **Ejemplo:** 4, 8, 16, 32.
* **Tipo de dato:** Numérico (entero).

### **`OS`**
* **Descripción:** Se refiere al sistema operativo preinstalado en la laptop.
* **Ejemplo:** Windows 10, macOS, Linux, Chrome OS.
* **Tipo de dato:** Categórico (texto).

### **`Weight`**
* **Descripción:** Indica el peso de la laptop, medido en kilogramos (kg).
* **Ejemplo:** 1.25, 2.30, 0.98.
* **Tipo de dato:** Numérico (flotante).

### **`Price_euros`**
* **Descripción:** El precio de venta de la laptop, expresado en euros (€). Esta es la variable objetivo que generalmente se busca predecir.
* **Ejemplo:** 999.00, 1499.99, 550.50.
* **Tipo de dato:** Numérico (flotante).

### **`ScreenW`**
* **Descripción:** Representa el ancho de la resolución de la pantalla en píxeles.
* **Ejemplo:** 1920 (Full HD), 2560 (QHD), 3840 (4K).
* **Tipo de dato:** Numérico (entero).

### **`ScreenH`**
* **Descripción:** Representa la altura de la resolución de la pantalla en píxeles.
* **Ejemplo:** 1080 (Full HD), 1440 (QHD), 2160 (4K).
* **Tipo de dato:** Numérico (entero).

### **`Touchscreen`**
* **Descripción:** Una variable binaria que indica si la laptop tiene una pantalla táctil.
* **Valores:** 0 (No), 1 (Sí).
* **Tipo de dato:** Numérico (entero/binario).

### **`IPSpanel`**
* **Descripción:** Una variable binaria que indica si la pantalla de la laptop utiliza tecnología de panel IPS (In-Plane Switching), que ofrece mejores ángulos de visión y reproducción de color.
* **Valores:** 0 (No), 1 (Sí).
* **Tipo de dato:** Numérico (entero/binario).

### **`RetinaDisplay`**
* **Descripción:** Una variable binaria que indica si la laptop tiene una pantalla Retina (término usado por Apple para pantallas de alta densidad de píxeles).
* **Valores:** 0 (No), 1 (Sí).
* **Tipo de dato:** Numérico (entero/binario).

### **`CPU_company`**
* **Descripción:** La compañía que fabrica el procesador (CPU) de la laptop.
* **Ejemplo:** Intel, AMD.
* **Tipo de dato:** Categórico (texto).

### **`CPU_freq`**
* **Descripción:** La frecuencia base del procesador (CPU) de la laptop, medida en Gigahertz (GHz). Una frecuencia más alta generalmente indica mayor velocidad de procesamiento.
* **Ejemplo:** 2.5, 3.0, 4.2.
* **Tipo de dato:** Numérico (flotante).

### **`PrimaryStorage`**
* **Descripción:** La capacidad del almacenamiento principal de la laptop, medida en Gigabytes (GB).
* **Ejemplo:** 256, 512, 1024.
* **Tipo de dato:** Numérico (entero).

### **`SecondaryStorage`**
* **Descripción:** La capacidad del almacenamiento secundario de la laptop, medido en Gigabytes (GB). Si la laptop no tiene un segundo disco, este valor puede ser 0.
* **Ejemplo:** 0, 128, 1000.
* **Tipo de dato:** Numérico (entero).

### **`PrimaryStorageType`**
* **Descripción:** El tipo de tecnología utilizada para el almacenamiento principal.
* **Ejemplo:** SSD (Solid State Drive), HDD (Hard Disk Drive), Hybrid.
* **Tipo de dato:** Categórico (texto).

### **`SecondaryStorageType`**
* **Descripción:** El tipo de tecnología utilizada para el almacenamiento secundario. Puede ser `NaN` o "No Aplica" si no hay almacenamiento secundario.
* **Ejemplo:** HDD, SSD, `NaN`.
* **Tipo de dato:** Categórico (texto).

### **`GPU_company`**
* **Descripción:** La compañía que fabrica la tarjeta gráfica (GPU) de la laptop.
* **Ejemplo:** Intel, Nvidia, AMD.
* **Tipo de dato:** Categórico (texto).
